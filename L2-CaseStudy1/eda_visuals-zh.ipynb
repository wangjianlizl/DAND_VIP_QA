```python
# 加载数据集
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
wine = pd.read_csv("./winequality_edited.csv")
```

```python
pd.plotting.scatter_matrix(wine[["fixed_acidity", "total_sulfur_dioxide", "pH", "alcohol"]],figsize=(10, 10))
plt.show()
```

```python
pd.plotting.scatter_matrix(wine[['volatile_acidity', 'residual_sugar',
       'pH', 'alcohol', "quality"]], figsize=(12, 12))
plt.show()
```
